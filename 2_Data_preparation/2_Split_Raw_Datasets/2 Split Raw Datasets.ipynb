```python
import pandas as pd

PATH_DATASETS = "../../2_Data_preparation/2_Split_Raw_Datasets/Datasets/"
PATH_OUTPUT = "output/"
PATH_STRENGTH = "../../2_Data_preparation/2_Split_Raw_Datasets/Datasets/Strength_AgeAdj_labeled.csv"
```

```python
def readCsvs(option):
    if option == 'features':
        datasets = []
        features = ['area', 'meancurv', 'thickness', 'volume']

        for feature in features:
            df = pd.read_csv(PATH_DATASETS + feature + '.csv', index_col=0)
            datasets.append({'name': feature, 'data': df})
        return datasets
    elif option == 'splits':
        datasets = []
        features = ['train', 'test', 'validate']

        for feature in features:
            df = pd.read_csv(PATH_DATASETS + feature + '.csv', index_col=0)
            datasets.append({'name': feature, 'data': df})
        return datasets
    else:
        print("the option is incorrect")
        
```

```python
def PipelineSplit(features, splits ):
    alr_splits = []
    train = []
    test = []
    validate = []
    
    for feature in features:
        for split in splits:
            df = build_dataset(split, feature)
            if split['name'] == 'train':
                train.append(df)
            elif split['name'] == 'test':
                test.append(df)
            elif split['name'] == 'validate':
                validate.append(df)
            else:
                print("error")
    alr_splits.append({'name': 'train', 'data': train})
    alr_splits.append({'name': 'test', 'data': test})
    alr_splits.append({'name': 'validate', 'data': validate})
    
    return alr_splits
    
    
```

```python
def build_dataset(Y_data, Full_data):
    X_data = pd.DataFrame(columns=Full_data['data'].columns)
    
    for index in Y_data['data'].index:
        r = Full_data['data'].loc[[index],:]
        X_data = pd.concat([X_data, r])
    
    df = {'name':Full_data['name'], 'data': X_data}
    return df
    
```

```python
# Reading the datasets 
Feature_Datasets = readCsvs('features') #Area,MeanCurv,Thickness,Volume
Splits_Datasets = readCsvs('splits') #train,test,validate
```

```python
# splits's structure:
# splits[0]['data']= splited datasets for train
# splits[0]['data'][0]['data'] = splited dataset of area for train
# splits[0]['data'][1]['data'] = splited dataset of meancurv for train
# splits[0]['data'][2]['data'] = splited dataset of thickness for train
# splits[0]['data'][3]['data'] = splited dataset of volume for train
# splits[1]['data']= splited datasets for test
#.
#.
#.
# splits[2]['data']= splited datasets for validate
#.
#.
#.
splits = PipelineSplit(Feature_Datasets, Splits_Datasets)
```

# saving the datasets X

```python
for split in splits:
    name_split = split['name']
    dataset = split['data']
    for feature in dataset:
        name_feature = feature['name']
        data = feature['data']
        data.to_csv(PATH_OUTPUT + 'X_' + name_feature + '_' + name_split + '.csv')
```
