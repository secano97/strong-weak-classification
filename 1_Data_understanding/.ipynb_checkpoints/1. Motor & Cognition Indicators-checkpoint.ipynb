```python
import pandas as pd
import scipy
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set(color_codes=True)
```

```python
dataset = pd.read_csv('unrestricted_federicocozza_7_19_2018_4_11_34.csv', sep=",", header=0, index_col=0)
```

## Motor

https://wiki.humanconnectome.org/display/PublicData/HCP+Data+Dictionary+Public-+Updated+for+the+1200+Subject+Release#HCPDataDictionaryPublic-Updatedforthe1200SubjectRelease-Category:Motor

Endurance: Endurance_Unadj, Endurance_AgeAdj<br/>
Locomotion: GaitSpeed_Comp<br/>
Dexterity: Dexterity_Unadj, Dexterity_AgeAdj<br/>
Strength: Strength_Unadj, Strength_AgeAdj

```python
variables1 = ['Endurance_Unadj', 'Endurance_AgeAdj', 'GaitSpeed_Comp', 'Dexterity_Unadj', 'Dexterity_AgeAdj', 'Strength_Unadj', 'Strength_AgeAdj']
```

```python
categories1 = ['Endurance', 'Endurance',
              'Locomotion',
              'Dexterity', 'Dexterity',
              'Strength', 'Strength']
```

```python
dataset_motor = dataset[variables1].copy()
```

```python
dataset_motor.head()
```

```python
i = -1
for var in variables1:
    
    i = i+1
    plt.subplots(figsize=(4, 4))
    ax = plt.axes()
    df = dataset_motor[[var]].copy()
    print(var)
    
    ## Delete the rows that have a NaN value
    if df[var].isnull().values.any():
        
        nan_values = df.isnull().values
        
        print("- Remove "+str(df[nan_values].shape[0])+" rows")
        df.dropna(inplace=True)
        df.reset_index(drop=True, inplace=True)
      
    print(df.describe())
    
    print("- Distribution")
    p = sns.distplot(df.values, ax=ax)
    
    x,y = p.get_lines()[0].get_data()
    cdf = scipy.integrate.cumtrapz(y, x, initial=0)
    
    #First Quartile
    first_q = np.abs(cdf-0.25).argmin()
    x_value = x[first_q]
    y_value = y[first_q]
    plt.vlines(x_value, 0, y_value, colors=['green'])
     
    #Second quartile
    second_q = np.abs(cdf-0.5).argmin()
    x_value = x[second_q]
    y_value = y[second_q]
    plt.vlines(x_value, 0, y_value, colors=['green'])
    
    #Third Quartile
    third_q = np.abs(cdf-0.75).argmin()
    x_value = x[third_q]
    y_value = y[third_q]
    plt.vlines(x_value, 0, y_value, colors=['green'])
    
    ax.set_title(categories1[i]+": "+var)
    plt.show()
    print("")
    
    #sns.set(style="whitegrid")
    #ax = sns.boxplot(x=df.values)
    #plt.show()
    print("")
    print("------")
```

```python
dataset_strength = dataset_motor[['Strength_AgeAdj']].copy()
```

```python
## Delete the rows that have a NaN value
if dataset_strength['Strength_AgeAdj'].isnull().values.any():

    nan_values = dataset_strength.isnull().values

    print("- Remove "+str(dataset_strength[nan_values].shape[0])+" rows")
    dataset_strength.dropna(inplace=True)
    
#dataset_strength.reset_index(drop=True, inplace=True)
```

```python
median = dataset_strength.median()[0]
print(median)
```

```python
dataset_strength.loc[dataset_strength['Strength_AgeAdj'] < median, 'class'] = -1
```

```python
dataset_strength.loc[dataset_strength['Strength_AgeAdj'] >= median, 'class'] = 1
```
