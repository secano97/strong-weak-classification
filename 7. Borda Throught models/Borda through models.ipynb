# <center> <b> <h1> Borda through models </h1></b></center>


```python
import pandas as pd
import numpy as np 

PATH_BORDA_RF = "../3.5 Modeling/3_Borda_method/output/borda classification.csv"
PATH_BORDA_SVM = "../4. Modeling_SVM/2. SVM RFE Borda/output/borda classification SVM Normalized.csv"
PATH_BORDA_LR = "../5. Modeling_logistic_regression/2. borda/output/borda classification logistic regresion.csv"
```

# 1. Preparing the data

```python
rf = pd.read_csv(PATH_BORDA_RF, index_col = 0)
svm = pd.read_csv(PATH_BORDA_SVM, index_col = 0)
lr = pd.read_csv(PATH_BORDA_LR, index_col = 0)
```

```python
rf
```

# 2. Create a df for the results 

```python
df_results = pd.DataFrame(columns=['Name', 'votes', 'rf', 'svm', 'lr'])
df_results['Name'] = rf.index.values
df_results['votes'] = 0
df_results['rf'] = 0
df_results['svm'] = 0
df_results['lr'] = 0
df_results.set_index('Name', inplace = True)
df_results.shape
```

```python
bordas = [rf, svm, lr]
model = 0
for borda in bordas:
    ROIs = borda.index.values
    position = 0
    for region in ROIs:
        votes = borda.at[region, 'votes']
        if votes < 500:
            if model == 0:
                df_results.at[region, 'rf'] = position
            elif model == 1:
                df_results.at[region, 'svm'] = position
            else:
                df_results.at[region, 'lr'] = position
                
            coordinate = df_results.at[region, 'votes']
            result = coordinate + position
            df_results.at[region, 'votes'] = result
        else:
            df_results.at[region, 'votes'] = 1000
        position += 1
    model += 1
```

```python
df_results.sort_values('votes', inplace = True)
```
